## Plots

```
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```
root_path = '/content/gdrive/My Drive/hyperbolic_nn'

rnn_in_embedds = np.loadtxt(root_path+"/2d_embedds/rand_in_hyperbolic_dim_2.txt",delimiter=',')
rnn_learn_embedds = np.loadtxt(root_path+"/2d_embedds/learned_hyperbolic_dim_2.txt",delimiter=',')

gensim_in_embedds = np.asarray(pickle.load(open(root_path+"/Gensim/2d_hyp_snli_embedds_gensim_1000_epoch", 'rb')))
gensim_learn_embedds = np.loadtxt(root_path+"/2d_embedds/2d_gensim_hyp_learned_embedds.txt",delimiter=',')

glv_in_embedds = np.asarray(pickle.load(open(root_path+"/2d_embedds/dim_2_epoch_50_init_pretrained", 'rb')))
glv_learn_embedds = np.loadtxt(root_path+"/2d_embedds/learned_poincare_glove_50_epoch_train_hyperbolic_dim_2.txt",delimiter=',')

sg_in_embedds = np.loadtxt(root_path+"/2d_embedds/2d_skipgram_hyp_trained_embedds.txt",delimiter=' ')
sg_learn_embedds = np.loadtxt(root_path+"/2d_embedds/learned_skipgram_train_hyperbolic_dim_2.txt",delimiter=',')
```

```
# Gensim

fig = plt.figure(figsize=(10,7))
ax1 = fig.add_subplot(111)

ax1.scatter(gensim_in_embedds[:,0], gensim_in_embedds[:,1],alpha=0.1,c='g',label='Gensim Pretrain')
ax1.scatter(gensim_learn_embedds[:,0], gensim_learn_embedds[:,1],alpha=0.5,c='b',label='Gensim Pretrain & Trained by Model')
ax1.scatter(rnn_learn_embedds [:,0], rnn_learn_embedds [:,1],alpha=0.5,c='r',label='Random In. & Trained by Model')
ax1.scatter(rnn_in_embedds[:,0], rnn_in_embedds[:,1],alpha=0.5,c='black',label='Random In.')


plt.legend(loc='lower left',fontsize=14)
ax1.spines['right'].set_color('none')
ax1.spines['top'].set_color('none')
plt.ylim([-1.4,1.4])
plt.xlim([-1.4,1.4])
plt.xlabel('Dimension 1',fontsize=18)
plt.ylabel('Dimension 2',fontsize=18)
plt.tick_params(labelsize=14)
plt.title("Hyperbolic 2-D Embeddings",fontsize=24)
#plt.savefig(root_path+'/Plots/Hyp_2d_Gensim_1000_Epochs.png')
plt.show()
```

```
# Poincare Glove

fig = plt.figure(figsize=(10,7))
ax1 = fig.add_subplot(111)

ax1.scatter(glv_learn_embedds[:,0], glv_learn_embedds[:,1],alpha=0.5,c='b',label='GloVe Pretrain & Trained by Model')
ax1.scatter(glv_in_embedds[:,0], glv_in_embedds[:,1],alpha=0.1,c='g',label='GloVe Pretrain')
ax1.scatter(rnn_learn_embedds [:,0], rnn_learn_embedds [:,1],alpha=0.5,c='r',label='Random In. & Trained by Model')
ax1.scatter(rnn_in_embedds[:,0], rnn_in_embedds[:,1],alpha=0.5,c='black',label='Random In.')


plt.legend(loc='lower right',fontsize=14)
ax1.spines['right'].set_color('none')
ax1.spines['top'].set_color('none')
# plt.ylim([-1.4,1.4])
# plt.xlim([-1.4,1.4])
plt.xlabel('Dimension 1',fontsize=18)
plt.ylabel('Dimension 2',fontsize=18)
plt.tick_params(labelsize=14)
plt.title("Hyperbolic 2-D Embeddings",fontsize=24)
plt.savefig(root_path+'/Plots/Hyp_2d_poincare_glove_50_Epochs.png')
plt.show()
```

```
# Skip Gram

fig = plt.figure(figsize=(10,7))
ax1 = fig.add_subplot(111)

ax1.scatter(sg_learn_embedds[:,0], sg_learn_embedds[:,1],alpha=0.5,c='b',label='Skip Gram Pretrain & Trained by Model')
ax1.scatter(sg_in_embedds[:,0], sg_in_embedds[:,1],alpha=0.1,c='g',label='Skip Gram Pretrain')
ax1.scatter(rnn_learn_embedds [:,0], rnn_learn_embedds [:,1],alpha=0.5,c='r',label='Random In. & Trained by Model')
ax1.scatter(rnn_in_embedds[:,0], rnn_in_embedds[:,1],alpha=0.5,c='black',label='Random In.')


plt.legend(loc='lower left',fontsize=14)
ax1.spines['right'].set_color('none')
ax1.spines['top'].set_color('none')
# plt.ylim([-1.4,1.4])
# plt.xlim([-1.4,1.4])
plt.xlabel('Dimension 1',fontsize=18)
plt.ylabel('Dimension 2',fontsize=18)
plt.tick_params(labelsize=14)
plt.title("Hyperbolic 2-D Embeddings",fontsize=24)
#plt.savefig(root_path+'/Plots/Hyp_2d_poincare_glove_50_Epochs.png')
plt.show()
```
