```
import pickle
import numpy as np
import matplotlib.pyplot as plt
import os

root_path = '/content/gdrive/My Drive/hyperbolic_nn'
snli_data_type = 'clean'
word_to_id_file_path = os.path.join(root_path, 'snli_dataset/', snli_data_type, 'word_to_id')
id_to_word_file_path = os.path.join(root_path, 'snli_dataset/', snli_data_type,'id_to_word')

pg_embeds_path = root_path + "/Poincare_glove/dim_2_epoch_50_init_pretrained.txt"

words = pickle.load(open(id_to_word_file_path, 'rb'))
words_set = set(words)

poincare_words = {}
in_em = []
out_em = []
with open(pg_embeds_path, 'r') as f:

    l = f.readline()

    while l:
        l = l.strip().split(' ')

        if l[0] in words_set:
            poincare_words[l[0]] = list(map(lambda x: float(x), l[1 :]))
        
        l = f.readline()

embedds = []
count_w = 0
count_t = 0
for w in words:
    count_t += 1
    if w in poincare_words:
        count_w += 1
        embedds.append(poincare_words[w])
        in_em.append(poincare_words[w])
    else:
        x = np.random.uniform(-0.001, 0.001, 2)
        embedds.append(np.random.uniform(-0.001, 0.001, 2))
        out_em.append(x)

print(count_w / count_t)
print(count_w)
print(count_t)
in_em = np.array(in_em)
out_em = np.array(out_em)
plt.scatter(in_em[:, 0], in_em[:, 1], c = 'r')
plt.scatter(out_em[:, 0], out_em[:, 1], c = 'b')
plt.show()
pickle.dump(embedds, open(root_path+'/2d_embedds/dim_2_epoch_50_init_pretrained', 'wb'))
```
